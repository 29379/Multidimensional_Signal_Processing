```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.random.randint(0, 255, size=(3,3))
plt.imshow(img, cmap='coolwarm')
```

```python
coords = [[0,0], [0,1], [0,2], [1,0], [1,1], [1,2], [2,0], [2,1], [2,2]]
coords = np.array(coords)

plt.scatter(coords[:,0], coords[:,1], c=img, cmap='coolwarm', s=3000)
plt.xlim(-1,3)
plt.ylim(-1,3)
plt.grid(ls=':')
```

```python
A = [[1, 0.5, 0],
     [0,1,0],
     [0,0,1]]
coords_ext = np.column_stack((coords, np.ones(9)))
coords_ext
```

```python
new_coords = coords_ext@A

fig, ax = plt.subplots(1,2, figsize=(10,5), sharex=True, sharey=True)
ax[0].scatter(coords[:,0], coords[:,1], c=img, cmap='coolwarm', s=1000)
ax[1].scatter(new_coords[:,0], new_coords[:,1], c=img, cmap='coolwarm', s=1000)

for a in ax:
    a.grid(ls=':')
```
